```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/ML/train.csv')
```

```python
train_df.head()
```

```python
test_df=pd.read_csv('/content/drive/MyDrive/ML/test.csv')
```

```python
test_df.head()
```

```python
X_train = train_df[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_train = train_df['SalePrice']
```

```python
sample = pd.read_csv('/content/drive/MyDrive/ML/sample_submission.csv')
sample.head()
```

```python
X_test = test_df[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_test = sample['SalePrice']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
output_df = pd.DataFrame({'Id': test_df['Id'], 'SalePrice': y_pred})
```

```python
output_df.to_csv('/content/drive/MyDrive/ML/predicted_prices.csv', index=False)
```
